```python
orders = pd.read_excel('sales.xls', 'Orders')
orders.head(3)
```

```python
app.layout = html.Div([
    
    dcc.Graph(
        id='scatter_chart',
        figure = {
            'data' : [
                go.Scatter(
                    x = orders.Sales,
                    y = orders.Profit,
                    mode = 'markers'
                    
                )
            ],
            'layout' : go.Layout(
                title = 'Sales and Profit',
                xaxis = {'title': 'Sales'},
                yaxis = {'title': 'Profit'},
                hovermode='closest'
            )
        }
    )
])
```

```python
if __name__ == '__main__':
    app.run_server(port = 4050)
```
